```python
#2
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x)
plt.plot(y)
```

```python

fig,ax1=plt.subplots()
fig=plt.figure(figsize=(12,6))

#x axis
ax1.set_xlabel('Numbers')
#bottom =False disables the ticks, label bottom disable x axis labels
ax1.tick_params(axis='x',bottom=False)
#set left y axis
ax1.set_ylabel('y label',color='red',size='x-large')

ax1.tick_params(axis='y',labelcolor='red',labelsize='large')

ax1.plot(x,
        color='red')

#set right y axis

#twinx sets the same x axis for plot plots
ax2=ax1.twinx()
ax2.set_ylabel('x label',color='blue',size='x-large')

ax2.tick_params(axis='y',labelcolor='blue',labelsize='large')

ax2.plot(y,
        color='blue')

plt.show()
```

```python
import pandas as pd

car_sales['Price'] = car_sales['Price'].replace('[\$\,]', '', regex=True).astype('int64')
```

```python
car_sales.head()
```

```python
#6
car_sales['Total Sales'] = car_sales['Price'].cumsum()
```

```python
#7
start_date = pd.to_datetime('today').date()
car_sales['Sale Date'] = pd.date_range(start=start_date, periods=len(car_sales))
car_sales.head()
```
